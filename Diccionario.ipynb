```python
'''Python Collections (Arrays)
There are four collection data types in the Python programming language:

List is a collection which is ordered and changeable. Allows duplicate members.
Tuple is a collection which is ordered and unchangeable. Allows duplicate members.
Set is a collection which is unordered, unchangeable*, and unindexed. No duplicate members.
Dictionary is a collection which is ordered** and changeable. No duplicate members.
*Set items are unchangeable, but you can remove and/or add items whenever you like.

**As of Python version 3.7, dictionaries are ordered. In Python 3.6 and earlier, dictionaries are unordered.

When choosing a collection type, it is useful to understand the properties of that type. Choosing the right type for a particular data set could mean retention of meaning, and, it could mean an increase in efficiency or security.
'''

#Diccionario. Par key:Value
d1 = {
  "Nombre": "Sara",
  "Edad": 27,
  "Documento": 1003882
}
print(d1)

#crear un diccionario en Python es usando dict()
print('Usando dict')
d2 = dict([
      ('Nombre', 'Sara'),
      ('Edad', 27),
      ('Documento', 1003882),
])
print(d2)
#Constructor
print('dict constructor')
d3 = dict(Nombre='Sara',
          Edad=27,
          Documento=1003882)
print(d3)
'''Propiedades:
1-Son dinámicos, pueden crecer o decrecer, se pueden añadir o eliminar elementos.
2-Son indexados, los elementos del diccionario son accesibles a través del key.
3-Y son anidados, un diccionario puede contener a otro diccionario en su campo value.'''

#Se puede acceder a sus elementos con [] o también con la función get()

print(d1['Nombre'])     #Sara
print(d1.get('Nombre')) #Sara

#modificar un elemento
d1['Direccion'] = "Calle 123"
print(d1)

#Iterar diccionario
print('Iterar diccionario')
# Imprime los key del diccionario
print('Imprime los key del diccionario')
for x in d1:
    print(x)

# Impreme los value del diccionario
print('Impreme los value del diccionario')
for x in d1:
    print(d1[x])
#key y value
# Imprime los key y value del diccionario
print('Impreme los key & value del diccionario')
for x, y in d1.items():
    print(x, y)

#Diccionarios anidados
print ('Diccionarios anidados')
anidado1 = {"a": 1, "b": 2}
anidado2 = {"a": 1, "b": 2}
d = {
  "anidado1" : anidado1,
  "anidado2" : anidado2
}
print(d)

#El método clear() elimina todo el contenido del diccionario.
print('Clear')
d = {'a': 1, 'b': 2}
d.clear()
print(d) #{}
#Get
print('get()')
#El método get() nos permite consultar el value para un key determinado
d = {'a': 1, 'b': 2}
print(d.get('a')) #1
print(d.get('z', 'No encontrado')) #No encontrado

print('items')
'''Def:
El método items() devuelve una lista con los keys y values del diccionario. 
Si se convierte en list se puede indexar como si de una lista normal se tratase, siendo los primeros elementos las key y los segundos los value.'''

d = {'a': 1, 'b': 2}
it = d.items()
print(it)             #dict_items([('a', 1), ('b', 2)])
print(list(it))       #[('a', 1), ('b', 2)]
print(list(it)[0][0]) #a

#keys:El método keys() devuelve una lista con todas las keys del diccionario.
print('keys()')
d = {'a': 1, 'b': 2}
k = d.keys()
print(k)       #dict_keys(['a', 'b'])
print(list(k)) #['a', 'b']
#Values()
print('values()')
d = {'a': 1, 'b': 2}
print(list(d.values())) #[1, 2]

#pop.El método pop() busca y elimina la key que se pasa como parámetro y devuelve su valor asociado. 
#Daría un error si se intenta eliminar una key que no existe.
print('pop()')
d = {'a': 1, 'b': 2}
d.pop('a')
print(d) #{'b': 2}

#update: update(<obj>)
'''El método update() se llama sobre un diccionario y tiene como entrada otro diccionario. 
Los value son actualizados y si alguna key del nuevo diccionario no esta, es añadida
'''
d1 = {'a': 1, 'b': 2}
d2 = {'a': 0, 'd': 400}
d1.update(d2)
print(d1)








```
