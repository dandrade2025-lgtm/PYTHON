```python
# Bucle for. Se utiliza para iterar sobre una secuencia (como una lista o una cadena)
print("----FOR-----")
numeros = [1, 2, 3, 4, 5]
for numero in numeros:
    print(numero)
print("----while-----")
# Bucle while. Se utiliza para repetir un bloque de código mientras una condición sea verdadera.
contador = 0
while contador < 5:
    print(contador)
    contador += 1
```

```python
#ejemplos for
print('texto al revés')
texto = "Python"
for i in texto[::-1]:
    print(i) #n,o,h,t,y,P

print('Itera la cadena saltándose elementos. Con [::2] vamos tomando un elemento si y otro no.')
texto = "Python"
for i in texto[::2]:
    print(i) #P,t,o
```

```python
#Uso de Range
for i in (0, 1, 2, 3, 4, 5):
    print(i) #0, 1, 2, 3, 4, 5
print('uso de range')
for i in range(6):
    print(i)
#El range() genera una secuencia de números que van desde 0 por defecto hasta el número que se pasa como parámetro menos 1.
#range(inicio, fin, salto)
print('range(inicio, fin, salto')
for i in range(5, 20, 2):
    print(i) #5,7,9,11,13,15,17,19

print('Se pueden generar también secuencias inversas, empezando por un número mayor y terminando en uno menor, pero para ello el salto deberá ser negativo.')
for i in range (5, 0, -1):
    print(i) #5,4,3,2,1
```

```python
#Break con bucles for
for i in range(5):
    print(i)
    break
    # No llega

cadena = 'Python'
for letra in cadena:
    if letra == 'h':
        print("Se encontró la h")
        break
    print(letra)


#Con while
x = 5
while True:
    x -= 1
    print(x)
    if x == 0:
        break
    print("Fin del bucle")
print('Bucles anidados, solo rompe uno')
for i in range(0, 4):
    for j in range(0, 4):
        break
        #Nunca se realiza más de una iteración
    # El break no afecta a este for
    print(i, j)

```

```python
#Cntinue: continue se salta todo el código restante en la iteración actual y
#vuelve al principio en el caso de que aún queden iteraciones por completar.
#La diferencia entre el break y continue es que el continue no rompe el bucle, si no que pasa a la siguiente iteración saltando el código pendiente.
print('for+continue')
cadena = 'Python'
for letra in cadena:
    if letra == 'P':
        continue
    print(letra)

print('while+continue')
x = 5
while x > 0:
    x -= 1
    if x == 3:
        continue
    print(x)
```

```python
#zip
a = [1, 2]
b = ["Uno", "Dos"]
c = zip(a, b)

print(list(c))

a = [1, 2]
b = ["Uno", "Dos"]
c = zip(a, b)

for numero, texto in zip(a, b):
    print("Número", numero, "Letra", texto)

numeros = [1, 2]
espanol = ["Uno", "Dos"]
ingles = ["One", "Two"]
frances = ["Un", "Deux"]
c = zip(numeros, espanol, ingles, frances)

for n, e, i, f in zip(numeros, espanol, ingles, frances):
    print(n, e, i, f)

#zipd diferentes longitudes
print('zipd diferentes longitudes')
numeros = [1, 2, 3, 4, 5]
espanol = ["Uno", "Dos"]

for n, e in zip(numeros, espanol):
    print(n, e)

#zip() con diccionarios
print('zip() con diccionarios')
esp = {'1': 'Uno', '2': 'Dos', '3': 'Tres'}
eng = {'1': 'One', '2': 'Two', '3': 'Three'}

for a, b in zip(esp, eng):
    print(a, b)

#zip() con diccionarios. hacemos uso de la función items, podemos acceder al key y value de cada elemento.
print('hacemos uso de la función items, podemos acceder al key y value de cada elemento.')
esp = {'1': 'Uno', '2': 'Dos', '3': 'Tres'}
eng = {'1': 'One', '2': 'Two', '3': 'Three'}

for (k1, v1), (k2, v2) in zip(esp.items(), eng.items()):
    print(k1, v1, v2)

# 1 Uno One
# 2 Dos Two
# 3 Tres Three


#Deshacer el zip()
print('Deshacer el zip()')
a = [1, 2, 3]
b = ["One", "Two", "Three"]
c = zip(a, b)

print(list(c))
print('Deshacemos')

c = [(1, 'One'), (2, 'Two'), (3, 'Three')]
a, b = zip(*c)

print(a)  # (1, 2, 3)
print(b)  # ('One', 'Two', 'Three')
```

```python
'''The else Statement
With the else statement we can run a block of code once when the condition no longer is true:'''
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```
