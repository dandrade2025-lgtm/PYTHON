```python
"""5 ejemplos de fácil a real para clase1. El más básico: dividir números"""
def dividir(a, b):
    try:
        resultado = a / b
    except ZeroDivisionError:
        print("Error: No puedes dividir entre 0")
        return None
    else:
        print("La división fue perfecta")
        return resultado
    finally:
        print("Operación terminada\n")

print(dividir(10, 2))  # Va por else
print(dividir(10, 0))  # Va por except
```

```python
"""Leer un número"""
try:
    edad = int(input("Dime tu edad: "))
except ValueError:
    print("Eso no es un número. Pon 18, 25, etc.")
else:
    if edad >= 18:
        print(f"Mayor de edad: {edad} años")
    else:
        print(f"Menor: {edad} años")
finally:
    print("Gracias por usar el programa")
```

```python
nombre_fichero = "datos.txt"

try:
    f = open(nombre_fichero, 'r')
except FileNotFoundError:
    print(f"El fichero {nombre_fichero} no existe")
except PermissionError:
    print("No tienes permisos para leerlo")
else:
    contenido = f.read()
    print(f"Fichero leído: {len(contenido)} caracteres")
    f.close()
finally:
    print("Intento de lectura finalizado")
```

```python
import requests

url = "https://api.github.com"

try:
    respuesta = requests.get(url, timeout=3)
    respuesta.raise_for_status()  # Lanza error si status 4xx/5xx
except requests.exceptions.Timeout:
    print("El servidor tardó mucho")
except requests.exceptions.RequestException:
    print("Error de conexión o URL mal")
else:
    print("Status OK:", respuesta.status_code)
    print("Datos:", respuesta.json()["current_user_url"])
finally:
    print("Petición finalizada")
```

```python
# SIN else - mal: mezclas código
try:
    num = int("42")
    print(f"El doble es {num * 2}")  # ¿Y si falla justo aquí?
except ValueError:
    print("No era número")

# CON else - bien: separas
try:
    num = int("42")  # Esto puede fallar
except ValueError:
    print("No era número")
else:
    print(f"El doble es {num * 2}")  # Esto solo si lo de arriba OK
```

```python
'''Dentro de funciones'''
#Ejemplos de try / except / else / finally en funciones 😊

#1. Dividir números con función

def dividir(a, b):
    try:
        resultado = a / b
    except ZeroDivisionError:
        print("Error: No puedes dividir entre 0")
        return None
    else:
        print("La división fue perfecta")
        return resultado
    finally:
        print("Operación terminada\n")

# Llamadas
print(dividir(10, 2))  # Va por else
print(dividir(10, 0))  # Va por except


#2. Leer edad del usuario

def pedir_edad():
    try:
        edad = int(input("Dime tu edad: "))
    except ValueError:
        print("Eso no es un número. Pon 18, 25, etc.")
    else:
        if edad >= 18:
            print(f"Mayor de edad: {edad} años")
        else:
            print(f"Menor: {edad} años")
    finally:
        print("Gracias por usar el programa")

# Llamada
pedir_edad()


#3. Abrir fichero

def leer_fichero(nombre):
    try:
        f = open(nombre, 'r')
    except FileNotFoundError:
        print(f"El fichero '{nombre}' no existe")
    except PermissionError:
        print("No tienes permisos para leerlo")
    else:
        contenido = f.read()
        print(f"Fichero leído: {len(contenido)} caracteres")
        f.close()
    finally:
        print("Intento de lectura finalizado")

# Llamadas
leer_fichero("datos.txt")
leer_fichero("no_existe.txt")


#4. Petición a API

import requests

def llamar_api(url):
    try:
        respuesta = requests.get(url, timeout=3)
        respuesta.raise_for_status()
    except requests.exceptions.Timeout:
        print("El servidor tardó mucho")
    except requests.exceptions.RequestException:
        print("Error de conexión o URL mal")
    else:
        print("Status OK:", respuesta.status_code)
        print("Datos:", respuesta.json()["current_user_url"])
    finally:
        print("Petición finalizada")

# Llamada
llamar_api("https://api.github.com")


#5. Conversión numérica

def convertir_a_numero(texto):
    try:
        num = int(texto)
    except ValueError:
        print("No era un número")
    else:
        print(f"El doble es {num * 2}")
    finally:
        print("Conversión intentada")

# Llamadas
convertir_a_numero("42")
convertir_a_numero("hola")
```
